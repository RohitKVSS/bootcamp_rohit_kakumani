```python
import sys
import os
import pandas as pd
from dotenv import load_dotenv
sys.path.append(os.path.abspath("../")) 
from src.cleaning import fill_missing_median, drop_missing, normalize_data

# Load .env for folder paths
load_dotenv()
RAW = os.getenv("DATA_DIR_RAW", "../data/raw")
PROC = os.getenv("DATA_DIR_PROCESSED", "../data/processed")
# Load dataset
file_path = os.path.join(RAW, "sample_data.csv")
df_orig = pd.read_csv(file_path, parse_dates=True)

print("Original Data:")
display(df_orig.head())
```

```python
df_filled = fill_missing_median(df_orig)

# Drop any remaining rows with missing values (optional)
df_dropped = drop_missing(df_filled)

# Normalize numeric columns
df_cleaned = normalize_data(df_dropped)

print("Cleaned Data:")
display(df_cleaned.head())
df_cleaned.to_csv(f'{PROC}/sample_data_cleaned.csv', index=False)
```

```python
print(f"Original shape: {df_orig.shape}")
print(f"Cleaned shape: {df_cleaned.shape}")

# Summary statistics
print("Original Summary:")
display(df_orig.describe())

print("Cleaned Summary:")
display(df_cleaned.describe())

# Missing values
print("Original NA counts:")
display(df_orig.isna().sum())

print("Cleaned NA counts:")
display(df_cleaned.isna().sum())
```
